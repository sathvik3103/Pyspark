```python
df_pyspark.na.drop().show() #by default this is how='any'. but we can chage it to how='all'
```

```python
df_pyspark.na.drop(how='any', thresh=2).show() #threshold is for number of non null values in a row
```

```python
df_pyspark.na.drop(how='any',subset=['Tax Per Product']).show() #to drop the whole records if there are null values from the specificed column
```

```python
df_pyspark.na.fill('Missing Values').show()
```

```python
df_pyspark.na.fill('Missing Values',['Tax Per Product', ' OrderID ']).show()
```

```python
from pyspark.ml.feature import Imputer
imputer = Imputer(
    inputCols=['Tax Per Product'],
    outputCols=["{}_imputed".format(c) for c in ['Tax Per Product']]
    ).setStrategy("mean")#median, mode
```

```python
imputer.fit(df_pyspark).transform(df_pyspark).show()
```

```python
df_pyspark.filter("`Tax Per Product`>7.0").show()
```

```python
df_pyspark.filter(df_pyspark['Tax Per Product']>7.0).show() #another way to write
```

```python
df_pyspark.filter("`Tax Per Product`>7.0").select([' Customer Name   ','Tax Per Product']).show()
```

```python
df_pyspark.filter((df_pyspark['Tax Per Product']>7.0) &
                  (df_pyspark['Product Category']=="Electronics")).show()
```

```python
df_pyspark.filter((df_pyspark['Tax Per Product']>7.0) |
                  (df_pyspark['Product Category']=="Electronics")).show() #OR
```

```python
df_pyspark.filter(~(df_pyspark['Tax Per Product']>7.0)).show()
```
